## Síndrome do impacto

Agora que você esta familiarizado com a terminologia e anatomia do ombro. Vou colocar a referência de um artigo que fala síndrome do impacto do ombro:

[The Painful Shoulder: Shoulder Impingement Syndrome](https://www.ncbi.nlm.nih.gov/pmc/articles/PMC3785027/)

Vou ressaltar alguns pontos deste artigo:

* Existem na verdade 3 tipos de pinçamento(shoulder impingment)

> Aside from subacromial impingment, other less common forms of shoulder impingement include subcoracoid impingment and Internal Impingment.

No entando até o momento quase tudo apresentado neste guia se refere ao pinçamento subacromial, que de longe é o mais comum e normalmente o tendão do musculo supraespinhal é comprido.

* Na grande maioria dos casos é possível resolver a condição com tratamento conservador
> Studies show that conservative management of shoulder impingement syndrome resolves the problem in 70-90% of patients
